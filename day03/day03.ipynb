```python
from math import prod

def find_number(i, j, matrix):
    """Finds the number at position i,j in the matrix"""
    try:
        # Go left until we hit the start of the number
        while matrix[i][j].isdigit() and j > 0:
            j -= 1

        # if at a non-digit, go right one step
        if not matrix[i][j].isdigit():
            j += 1
        
        number = []
        while j < len(matrix[0]) and matrix[i][j].isdigit():
            number.append(matrix[i][j])
            j += 1

        if len(number) > 0:
            return int(''.join(number))
        else:
            return None
    except IndexError:
        print(i,j)
        raise
    
        

def part_two(matrix):
    valid_partnos = []

    for ix, row in enumerate(matrix):
        
        for iy, col in enumerate(row):
            gears = []
            if col == '*':
                for i,j in adjacent_positions(ix,iy,matrix):
                    if matrix[i][j].isdigit():
                        #print('isdigit', i,j,matrix[i][j])
                        adj_number = find_number(i,j,matrix)
                        #print('adj_number', adj_number)
                        if adj_number is not None:
                            gears.append(adj_number)
                #print(ix,iy, gears)
                gear_set = set(gears)
                if len(gear_set) == 2:
                    valid_partnos.append(prod(gear_set))

    return valid_partnos
```
