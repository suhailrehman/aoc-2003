```python
def parse_game(line):
    gameid = line.split(':')[0].split(' ')[1]

    games = line.split(':')[1].split(';')

    output_games = []
    # split each game into a dict by color and number
    for game in games:
        game_dict = {}
        for grab in game.split(','):
            #print(grab)
            number, color = grab.split()
            game_dict[color.strip()] = int(number.strip())
        output_games.append(game_dict)
    return gameid, output_games
```

```python
sample_input = read_input('input.txt', parse_game)
#full_input = read_input('input.txt')
```

```python
from collections import Counter
valid_configurations = {'red': 12, 'green': 13, 'blue': 14}

def part_one(sample_input):
    valid_games = []
    for game in sample_input:
        valid = True
        for grab in game[1]:
            revealed_colors = Counter()
            for color, number in grab.items():
                revealed_colors[color] += number
                if revealed_colors[color] > valid_configurations[color]:
                    valid = False
                    break
        if valid:
            valid_games.append(game[0])
    return valid_games
```

```python
sum([int(x) for x in part_one(sample_input)])
```
